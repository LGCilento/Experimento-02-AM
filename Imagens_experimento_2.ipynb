```
fig1, ax = plt.subplots() ######-----Curva ROC -----#########
fig1.set_figheight(7)
fig1.set_figwidth(15)

k = [1]

ax.plot(fp_rate_list_normal, tp_rate_list_normal,'r^', label='k-NN default')
ax.annotate("default", (fp_rate_list_normal[0], tp_rate_list_normal[0]), fontsize=15)

ax.plot(fp_rate_list_lvq1, tp_rate_list_lvq1,'go', label='lvq1')
for i, txt in enumerate(n):
    ax.annotate(txt, (fp_rate_list_lvq1[i], tp_rate_list_lvq1[i]), fontsize=15)

ax.plot(fp_rate_list_lvq2, tp_rate_list_lvq2,'kx',label='lvq2')
for i, txt in enumerate(n):
    ax.annotate(txt, (fp_rate_list_lvq2[i], tp_rate_list_lvq2[i]), fontsize=15)

ax.plot(fp_rate_list_lvq3, tp_rate_list_lvq3,'cs',label='lvq3')
for i, txt in enumerate(n):
    ax.annotate(txt, (fp_rate_list_lvq3[i], tp_rate_list_lvq3[i]), fontsize=15)
                
ax.set(xlabel='FP_rate', ylabel='TP_rate',title='ROC curve for K = 1')
ax.grid()
ax.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1)
fig1.savefig('image1.png', bbox_inches='tight')
plt.show()
```

```
fig1, ax = plt.subplots() ######-----Curva Evolução TP -----#########
fig1.set_figheight(7)
fig1.set_figwidth(15)

#ax.plot(k, tp_rate_list_adaptative,marker='^',color='r', label='k-NN default')
ax.plot(n, tp_rate_list_lvq1,marker='o',color='g', label='lvq1')
ax.plot(n, tp_rate_list_lvq2,marker='x',color='k',label='lvq2',markersize=15)
ax.plot(n, tp_rate_list_lvq3,marker='s',color='c',label='lvq3')


ax.set(xlabel='N_Prototypes', ylabel='TP_rate',title='TP_RATE for K = 1')
ax.grid()
ax.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1)
fig1.savefig('image2.png', bbox_inches='tight')
plt.show()
```

```
fig1, ax = plt.subplots() ######-----Curva Evolução FP -----#########
fig1.set_figheight(7)
fig1.set_figwidth(15)

ax.plot(n, fp_rate_list_lvq1,marker='o',color='g', label='lvq1')
ax.plot(n, fp_rate_list_lvq2,marker='x',color='k',label='lvq2',markersize=15)
ax.plot(n, fp_rate_list_lvq3,marker='s',color='c',label='lvq3')


ax.set(xlabel='N_Prototypes', ylabel='FP_rate',title='FP_RATE for K = 1')
ax.grid()
ax.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1)
fig1.savefig('image3.png', bbox_inches='tight')
plt.show()
```
